# Building learning models

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
X = Data[['Age','Years of experience','Work Location','Job']]
y = Data[['Salary (in USD)']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
lin = LinearRegression()
lin.fit(X_train, y_train)

lin_y_pred = lin.predict(X_test)

print('Mean Squared Error: ',mean_squared_error(y_test, lin_y_pred))
print('R2 score: ',r2_score(y_test, lin_y_pred))
```

# Gradient Boosting regressor

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr = GradientBoostingRegressor(random_state = 1)
gbr.fit(X_train, y_train.values.ravel())

gbr_y_pred = gbr.predict(X_test)

print('Mean Squared Error: ',mean_squared_error(y_test, gbr_y_pred))
print('R2 score: ',r2_score(y_test, gbr_y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gbr = GradientBoostingRegressor(random_state = 1)

param_grid = {
    'n_estimators': [100],
    'learning_rate': [0.1, 0.01, 1],
    'max_depth': [3,4,5]
}

grid_search = GridSearchCV(gbr, param_grid, cv=4, scoring=['r2','neg_mean_squared_error'], refit = 'neg_mean_squared_error')
grid_search.fit(X_train, y_train.values.ravel())

best_model = grid_search.best_estimator_
best_params = grid_search.best_params_

gbr_y_pred = best_model.predict(X_test)

print("Best Hyperparameters:", best_params)
print("Validation MSE:", mean_squared_error(y_test, gbr_y_pred))
print("R squared error:",r2_score(y_test, gbr_y_pred))
```

```python
import dash
from dash import html
from dash import dcc
from dash import Dash
from dash.dependencies import Input, Output, State
```

```python
def create_line_chart(line_x='Years of experience', line_y='Salary (in USD)'):
    y = []
    for i in np.sort(Data[line_x].unique()):
        n = Data[Data[line_x]==i][line_y].mean()
        y.append(n)
    layout = go.Layout(title = '{} vs {}'.format(line_x.capitalize(), line_y.capitalize()))
    if line_x == 'Work Location':
        line_fig = go.Figure(data = [go.Scatter(x=np.sort(loc_le.inverse_transform(Data[line_x].unique())), y=y)], layout=layout)
    else:
        line_fig = go.Figure(data = [go.Scatter(x=np.sort(Data[line_x].unique()), y=y)], layout=layout)
    return line_fig
line_x = dcc.Dropdown(id='line_x', options=['Age','Years of experience','Work Location'], value='Years of experience',
                      clearable=False,style={'display':'inline_block','width':'50%'})
line_y = dcc.Dropdown(id='line_y', options=['Age','Years of experience','Salary (in USD)'], value='Salary (in USD)',
                      clearable=False,style={'display':'inline_block','width':'50%'})

def create_bar_chart(bar_x='Years of experience', bar_y='Salary (in USD)'):
    y = []
    for i in np.sort(Data[bar_x].unique()):
        n = Data[Data[bar_x]==i][bar_y].mean()
        y.append(n)
    layout = go.Layout(title = '{} vs {}'.format(bar_x.capitalize(), bar_y.capitalize()))
    if bar_x == 'Work Location':
        bar_fig = go.Figure(data = go.Bar(x=np.sort(loc_le.inverse_transform(Data[bar_x].unique())), y=y), layout=layout)
    else:
        bar_fig = go.Figure(data = go.Bar(x=np.sort(Data[bar_x].unique()), y=y), layout=layout)
    return bar_fig
bar_x = dcc.Dropdown(id='bar_x', options=['Age','Years of experience','Work Location'], value='Years of experience',
                     clearable=False,style={'display':'inline_block','width':'50%'})
bar_y = dcc.Dropdown(id='bar_y', options=['Age','Years of experience','Salary (in USD)'], value='Salary (in USD)',
                     clearable=False,style={'display':'inline_block','width':'50%'})

def create_pie_chart(sectors='Work Location'):
    if sectors == 'Work Location':
        pie_fig = px.pie(Data, names = loc_le.inverse_transform(Data[sectors]),
                    title = 'Percentage of employees in each {}'.format(sectors.capitalize()))
    elif sectors == 'Job':
        pie_fig = px.pie(Data, names = job_le.inverse_transform(Data[sectors]),
                    title = 'Percentage of employees in each {}'.format(sectors.capitalize()))
    else:
        pie_fig = px.pie(Data, names = sectors,
                    title = 'Percentage of employees in each {}'.format(sectors.capitalize()))
    return pie_fig
sectors = dcc.Dropdown(id='sectors', options=['Age','Years of experience','Work Location','Job'], value='Work Location',
                       clearable=False,style={'display':'inline_block','width':'40%'})

def create_job_line_chart(job_line_x='Years of experience', job_line_y='Salary (in USD)', job_line='TEACHER'):
    layout = go.Layout(title = '{} vs {}(Job:{})'.format(job_line_x.capitalize(), job_line_y.capitalize(),job_line))
    job_line = int(job_le.transform([job_line]))
    data=Data[Data['Job']==job_line]
    y = []
    for i in np.sort(data[job_line_x].unique()):
        n = data[data[job_line_x]==i][job_line_y].mean()
        y.append(n)
    if job_line_x == 'Work Location':
        job_line_fig = go.Figure(data = [go.Scatter(x=np.sort(loc_le.inverse_transform(data[job_line_x].unique())), y=y)], layout=layout)
    else:
        job_line_fig = go.Figure(data = [go.Scatter(x=np.sort(data[job_line_x].unique()), y=y)], layout=layout)
    return job_line_fig
job_line_x = dcc.Dropdown(id='job_line_x', options=['Age','Years of experience','Work Location'], value='Years of experience',
                          clearable=False,style={'display':'inline_block','width':'50%'})
job_line_y = dcc.Dropdown(id='job_line_y', options=['Age','Years of experience','Salary (in USD)'], value='Salary (in USD)',
                          clearable=False,style={'display':'inline_block','width':'50%'})
job_line = dcc.Dropdown(id='job_line', options=np.sort(job_le.inverse_transform(Data['Job'].unique())), value='TEACHER',
                        clearable=False,style={'display':'inline_block','width':'50%'})

def create_job_bar_chart(job_bar_x='Years of experience', job_bar_y='Salary (in USD)', job_bar='TEACHER'):
    layout = go.Layout(title = '{} vs {}(Job:{})'.format(job_bar_x.capitalize(), job_bar_y.capitalize(),job_bar))
    job_bar = int(job_le.transform([job_bar]))
    data=Data[Data['Job']==job_bar]
    y = []
    for i in np.sort(data[job_bar_x].unique()):
        n = data[data[job_bar_x]==i][job_bar_y].mean()
        y.append(n)
    if job_bar_x == 'Work Location':
        job_bar_fig = go.Figure(data = [go.Bar(x=np.sort(loc_le.inverse_transform(data[job_bar_x].unique())), y=y)], layout=layout)
    else:
        job_bar_fig = go.Figure(data = [go.Bar(x=np.sort(data[job_bar_x].unique()), y=y)], layout=layout)
    return job_bar_fig
job_bar_x = dcc.Dropdown(id='job_bar_x', options=['Age','Years of experience','Work Location'], value='Years of experience',
                         clearable=False,style={'display':'inline_block','width':'50%'})
job_bar_y = dcc.Dropdown(id='job_bar_y', options=['Age','Years of experience','Salary (in USD)'], value='Salary (in USD)',
                         clearable=False,style={'display':'inline_block','width':'50%'})
job_bar = dcc.Dropdown(id='job_bar', options=np.sort(job_le.inverse_transform(Data['Job'].unique())), value='TEACHER',
                       clearable=False,style={'display':'inline_block','width':'50%'})
```

```python
app = Dash(title = 'HR SALARY DASHBOARD')

app.layout = html.Div(
    children=[
        html.H1('HR Salary Analysis',style={'text-align':'center'}),
        html.Br(),
        html.Div(
            children=[
                html.H2('Line plot',style={'text-align':'center'}),
                html.Label('X-axis:'),
                line_x,
                html.Label('Y-axis:'),
                line_y,
                dcc.Graph(id='line_chart', figure=create_line_chart())
            ],
            style={'display':'inline-block','width':'49%'}),
        html.Div(
            children=[
                html.H2('Bar plot',style={'text-align':'center'}),
                html.Label('X-axis:'),
                bar_x,
                html.Label('Y-axis:'),
                bar_y,
                dcc.Graph(id='bar_chart', figure=create_bar_chart())
            ],
            style={'display':'inline-block','width':'49%'}
        ),
        html.Br(),
        html.Div(
            children=[
                html.H2('Pie chart',style={'text-align':'center'}),
                html.Label('Categorize on:'),
                sectors,
                dcc.Graph(id='pie_chart', figure=create_pie_chart())
            ]
        ),
        html.Br(),
        html.Div(
            children=[
                html.H2('Line plot based on Job',style={'text-align':'center'}),
                html.Label('X-axis:'),
                job_line_x,
                html.Label('Y-axis:'),
                job_line_y,
                html.Label('Job:'),
                job_line,
                dcc.Graph(id='job_line_chart', figure=create_job_line_chart())
            ],
            style={'display':'inline-block','width':'49%'}
        ),
        html.Div(
            children=[
                html.H2('Bar plot based on Job',style={'text-align':'center'}),
                html.Label('X-axis:'),
                job_bar_x,
                html.Label('Y-axis:'),
                job_bar_y,
                html.Label('Job:'),
                job_bar,
                dcc.Graph(id='job_bar_chart', figure=create_job_bar_chart())
            ],
            style={'display':'inline-block','width':'49%'}
        ),
        html.Br(),
        html.H1('Salary Prediction',style={'text-align':'center'}),
        html.Br(),
        html.Div(
            children=[
                html.Label('Select Age'),
                dcc.Dropdown(
                    id='age',options=np.sort(Data['Age'].unique()),
                    style={'display':'inline_block','width':'50%'}),
                html.Br(),
                html.Label('Select Years of experience'),
                dcc.Dropdown(
                    id='years',options=np.sort(Data['Years of experience'].unique()),
                    style={'display':'inline_block','width':'50%'}),
                html.Br(),
                html.Label('Select Location'),
                dcc.Dropdown(
                    id='location',options=[{'label':loc_le.inverse_transform([location]),'value':location}
                                          for location in np.sort(Data['Work Location'].unique())],
                    style={'display':'inline_block','width':'50%'}),
                html.Br(),
                html.Label('Select Job'),
                dcc.Dropdown(
                    id='job',options=[{'label':job_le.inverse_transform([job]),'value':job}
                                     for job in np.sort(Data['Job'].unique())],
                    style={'display':'inline_block','width':'50%'})
            ]),
        html.Br(),
        html.Button('Submit', id='submit-button', n_clicks=0),
        html.Div(id='prediction-output')
    ],
)

@app.callback(Output('line_chart','figure'),[Input('line_x','value'),
                                             Input('line_y','value')])
def update_line_chart(line_x, line_y):
    return create_line_chart(line_x, line_y)

@app.callback(Output('bar_chart','figure'),[Input('bar_x','value'),
                                            Input('bar_y','value')])
def update_bar_chart(bar_x, bar_y):
    return create_bar_chart(bar_x, bar_y)

@app.callback(Output('pie_chart','figure'),[Input('sectors','value')])
def updata_pie_chart(sectors):
    return create_pie_chart(sectors)

@app.callback(Output('job_line_chart','figure'),
              [Input('job_line_x','value'),
               Input('job_line_y','value'),
               Input('job_line','value')])
def update_job_line_chart(job_line_x, job_line_y, job_line):
    return create_job_line_chart(job_line_x, job_line_y, job_line)

@app.callback(Output('job_bar_chart','figure'),
              [Input('job_bar_x','value'),
               Input('job_bar_y','value'),
               Input('job_bar','value')])
def update_job_bar_chart(job_bar_x, job_bar_y, job_bar):
    return create_job_bar_chart(job_bar_x, job_bar_y, job_bar)

@app.callback(Output('prediction-output','children'),
              [Input('submit-button','n_clicks')],
              [State('age','value'),
               State('years','value'),
               State('location','value'),
               State('job','value')])
def predict_salary(n_clicks, age, years, location, job):
    if n_clicks > 0:
        if (age < years) | (age-years < 20):
            return html.H2('Enter valid age and years of experience')
        else:
            prediction = Best_model.predict([[age, years, location, job]])
            return html.H2(f'Predicted salary : ${prediction[0]:,.2f}',style={'text-align':'center'})
    else:
        return html.H2('Click submit to get the prediction',style={'text-align':'center'})
    
if __name__ == '__main__':
    app.run()
```
