# Cleaning and Sanitizing

```python
sum(Data['First Name'].isnull() & Data['Last Name'].isnull())
```

```python
Data.dropna(subset=['First Name', 'Last Name'], thresh = 1, inplace = True)
```

```python
Data[Data['Last Name'].isnull()]
```

```python
Data['Name'] = Data['First Name']+' '+Data['Last Name']
```

```python
Data.head()
```

# Removing unnecessary columns

```python
Data.drop(columns = ['First Name', 'Last Name', 'Mid Init'], inplace = True)
```

```python
Data.drop(columns = ['Fiscal Year', 'Payroll Number', 'Pay Basis', 'Leave Status as of June 30'], inplace = True)
```

```python
Data.head()
```

```python
Data = Data[['Agency Name', 'Name', 'Age', 'Joined Date', 'Work Location Borough', 'Job Description', 'Base Salary']]
```

```python
Data.head()
```

# Renaming the columns

```python
Data.rename(columns = {'Work Location Borough':'Work Location', 'Job Description':'Job', 'Base Salary':'Salary'}, inplace = True)
```

```python
Data.head()
```

```python
Data['Joined Date'] = pd.to_datetime(Data['Joined Date'])
```

```python
Data['Joined Year'] = Data['Joined Date'].dt.year
```

```python
Data['Years of experience'] = 2022 - Data['Joined Year']
```

```python
Data.head()
```

```python
Data.drop(columns = ['Joined Date', 'Joined Year'], inplace = True)
```

```python
Data = Data[['Agency Name', 'Name', 'Age', 'Years of experience', 'Work Location', 'Job', 'Salary']]
```

```python
Data.head()
```

```python
Data['Salary'] = Data['Salary'].astype(int)
```

```python
Data.info()
```

# Dropping null values

```python
Data.dropna(inplace = True)
```
